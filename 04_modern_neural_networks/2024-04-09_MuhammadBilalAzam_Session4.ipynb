```python
import string
import re
from operator import itemgetter

# Part 1: Tokenization
wdict = {}
total_word_count = 0

with open('Life_On_The_Mississippi.txt', 'r', encoding='utf-8-sig') as L:  # Fix BOM issue
    for line in L:
        line = line.lower()  # Normalize to lowercase
        line = line.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
        line = re.sub(r'[^\x00-\x7f]', '', line)  # Remove non-ASCII characters

        words = line.split()
        total_word_count += len(words)

        for word in words:
            wdict[word] = wdict.get(word, 0) + 1

# Part 2: Sorting and Analysis
sorted_wdict = sorted(wdict.items(), key=itemgetter(1), reverse=True)
top_100_words = sorted_wdict[:100]

# Calculate total occurrences for top 90% of words
cumulative_count = 0
for word, frequency in sorted_wdict:
    cumulative_count += frequency
    if cumulative_count >= total_word_count * 0.9:
        break

print("Top 100 most frequent word tokens:")
for word, frequency in top_100_words:
    print(word, frequency)

print("\nNumber of distinct words making up the top 90% of word occurrences:", len(sorted_wdict[:sorted_wdict.index((word, frequency))]))
```
